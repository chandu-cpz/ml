# Import Lib's and Dataset

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("datasets/insurance.csv")
df.head()
```

```python
df.shape
```

# Univariate Analysis

```python
sns.distplot(df.age)
```

```python
sns.displot(df.bmi)
```

```python
plt.pie(df.smoker.value_counts(),explode=[0,0.1],labels=df.smoker.value_counts().index,shadow=True,autopct='%1.1f%%',colors=['pink','skyblue'])
plt.title("Smokers")
plt.show()
```

```python
sns.barplot(x = df.region.value_counts().index,y = df.region.value_counts())
```

# Bivariate Analysis

```python
sns.lineplot(x = df.age , y = df.charges)
```

```python
sns.scatterplot(x = df.age , y = df.charges)
```

# Multivariate Analysis

```python
sns.pairplot(df)
```
